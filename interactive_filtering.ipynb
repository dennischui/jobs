```python
from pathlib import Path
link = Path(r'./sample_pages/Culture Amp response text copy.htm')

with open(link, "r", encoding="utf-8") as f:
                    html_content = f.read()
```

```python
from google import genai
from google.genai import types

client = genai.Client()

response = client.models.generate_content(
    model="gemini-3-flash-preview",
    config=types.GenerateContentConfig(
        system_instruction="""You are an HTML parsing engine.

        You are given raw HTML as plain text, not a rendered webpage.
        You do not execute JavaScript, load external resources, or infer missing elements.
        You only use the tags, attributes, and text explicitly present in the HTML.

        Treat HTML tags as structural markers.
        Use nesting, tag names, ids, classes, and attributes to understand hierarchy.

        Rules:
        - Ignore <script>, <style>, and HTML comments unless explicitly requested.
        - Do not hallucinate content that is not present.
        - If data is missing, return null.
        - If multiple matches exist, return all of them.
        - Preserve original text exactly (no rewriting unless asked).
        - Whitespace may be normalized unless exact formatting is requested.

        When extracting data:
        - Prefer semantic tags (<article>, <section>, <nav>, <header>, <footer>) when available.
        - Use class and id attributes to disambiguate elements.
        - If structure is ambiguous, explain the ambiguity briefly.

        Output format:
        - Always return valid JSON.
        - Do not include explanatory text unless explicitly requested.
        - Match the requested schema exactly.

        Assume the HTML may be malformed.
        Do your best to interpret structure based only on the text provided.

        """),
    contents=html_content
)

print(response.text)
```
